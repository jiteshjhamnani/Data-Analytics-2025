```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.experimental import enable_iterative_imputer
from sklearn.model_selection import train_test_split
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python

train_df = pd.read_csv("https://github.com/jiteshjhamnani/Data-Analytics-2025/raw/refs/heads/main/hacktrain.csv")
test_df = pd.read_csv("https://github.com/jiteshjhamnani/Data-Analytics-2025/raw/refs/heads/main/hacktest.csv")

test_id = test_df['ID']

train_df.drop(columns=['Unnamed: 0'], inplace=True)
test_df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python

le = LabelEncoder()
train_df['class'] = le.fit_transform(train_df['class'])

X_train = train_df.drop(columns=['ID', 'class'])
y_train = train_df['class']
X_test = test_df.drop(columns=['ID'])

pipeline = Pipeline([
    ('imputer', IterativeImputer(random_state=42)),
    ('scaler', StandardScaler()),
    ('logreg', LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42))
])

pipeline.fit(X_train, y_train)

test_preds = pipeline.predict(X_test)

predicted_labels = le.inverse_transform(test_preds)

submission = pd.DataFrame({
    'ID': test_id,
    'class': predicted_labels
})
```

```python
print(submission.head())

submission.to_csv("submission.csv", index=False)

from google.colab import files
files.download('submission.csv')
```
