```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import KNNImputer
from sklearn.metrics import f1_score
from imblearn.over_sampling import SMOTENC
from sklearn.ensemble import GradientBoostingClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical

df = pd.read_csv("/content/Train_Data.csv")

df.dropna(subset=["age_group"], inplace=True)

le = LabelEncoder()
df['age_group'] = le.fit_transform(df['age_group'])

df['BMI_cat'] = pd.cut(df['BMXBMI'], bins=[0, 18.5, 25, 30, 100], labels=[0, 1, 2, 3])
df['Glucose_band'] = pd.cut(df['LBXGLU'], bins=[0, 100, 126, 500], labels=[0, 1, 2])

df.drop(columns=['SEQN'], inplace=True)
df['BMI_cat'] = df['BMI_cat'].astype(float)
df['Glucose_band'] = df['Glucose_band'].astype(float)

X = df.drop(columns=['age_group'])
y = df['age_group']

imputer = KNNImputer(n_neighbors=3)
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

cat_features = ['RIAGENDR', 'PAQ605', 'DIQ010', 'BMI_cat', 'Glucose_band']
cat_indices = [X.columns.get_loc(col) for col in cat_features]

smote_nc = SMOTENC(categorical_features=cat_indices, random_state=42)
X_resampled, y_resampled = smote_nc.fit_resample(X_imputed, y)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_resampled, test_size=0.2, random_state=42)

gb_clf = GradientBoostingClassifier(random_state=42)
gb_clf.fit(X_train, y_train)
gb_preds = gb_clf.predict(X_test)
print("Gradient Boosting F1 Score:", f1_score(y_test, gb_preds, average='weighted'))

y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)

model = Sequential([
    Dense(64, input_dim=X_train.shape[1], activation='relu'),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(2, activation='softmax')
])
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train_cat, epochs=30, batch_size=32, verbose=0, validation_split=0.1)

nn_preds = model.predict(X_test)
nn_preds_classes = np.argmax(nn_preds, axis=1)
print("Neural Network F1 Score:", f1_score(y_test, nn_preds_classes, average='weighted'))
```

```python

test_df['BMI_cat'] = pd.cut(test_df['BMXBMI'], bins=[0, 18.5, 25, 30, 100], labels=[0, 1, 2, 3])
test_df['Glucose_band'] = pd.cut(test_df['LBXGLU'], bins=[0, 100, 126, 500], labels=[0, 1, 2])
test_df['BMI_cat'] = test_df['BMI_cat'].astype(float)
test_df['Glucose_band'] = test_df['Glucose_band'].astype(float)

seqn_ids = test_df['SEQN'] if 'SEQN' in test_df.columns else np.arange(len(test_df))

test_df = test_df.drop(columns=['SEQN'], errors='ignore')

test_df_imputed = pd.DataFrame(imputer.transform(test_df), columns=test_df.columns)

test_scaled = scaler.transform(test_df_imputed)

nn_test_preds = model.predict(test_scaled)
nn_test_preds_class = np.argmax(nn_test_preds, axis=1)
submission = pd.DataFrame({
    'SEQN': seqn_ids,
    'age_group': nn_test_preds_class
})
submission.to_csv("submission.csv", index=False)
print(submission.shape)
```
